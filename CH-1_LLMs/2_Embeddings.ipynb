# Embedding Vectors Comparison

This section compares embedding vectors generated using two different OpenAI models:
- **text-embedding-3-small**: A smaller, faster model (Cell 1)
- **text-embedding-3-large**: A larger, more detailed model (Cell 2)

Both models generate vector embeddings for the same input string "I am a test string", allowing us to observe differences in:
- Vector dimensions (size)
- Computational efficiency
- Semantic representation quality

### **text-embedding-3-small**: A smaller, faster model 

```python
from openai import OpenAI
client = OpenAI()

response = client.embeddings.create(
    input="I am a test string",
    model="text-embedding-3-small" 
)

vector=response.data[0].embedding
print(vector[:10])
print("Size of vector: ", len(vector))
```

### **text-embedding-3-large**: A larger, more detailed model

```python
from openai import OpenAI
client = OpenAI()

response = client.embeddings.create(
    input="I am a test string",
    model="text-embedding-3-large" 
)

vector=response.data[0].embedding
print(vector[:10])
print("Size of vector: ", len(vector))
```
