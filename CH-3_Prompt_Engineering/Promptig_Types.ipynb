# Prompt Engineering: Types of Prompting

## 1. Zero-shot Prompting

Zero-shot prompting is when you ask a language model to perform a task without providing any examples. This relies on the model's pre-trained knowledge.

## Benefits:

- Fast and simple
- No need for labeled examples
- Useful for general tasks

```python
# Example: Zero-shot prompt (pseudo-code, replace with actual API call as needed)
zero_shot_prompt = "Translate the following English sentence to French: 'How are you?'"
# response = language_model_api(prompt=zero_shot_prompt)
response = "Comment ça va ?"  # Example output
print("Prompt:", zero_shot_prompt)
print("Response:", response)
```

# 2. One-shot Prompting

One-shot prompting provides the model with a single example to help it understand the task.

## Benefits:

- Gives the model context
- Useful when task is ambiguous without an example

```python
# Example: One-shot prompt (pseudo-code)
one_shot_prompt = (
    "Translate the following English sentence to French.\n"
    "Example: 'Good morning' -> 'Bonjour'\n"
    "Now, translate: 'How are you?'"
)
# response = language_model_api(prompt=one_shot_prompt)
response = "Comment ça va ?"  # Example output
print("Prompt:\n", one_shot_prompt)
print("Response:", response)
```

# 3. Few-shot Prompting

Few-shot prompting provides the model with several examples to clarify the task and desired output format.

## Benefits:

- Improves accuracy for complex tasks
- Helps the model generalize from examples

```python
# Example: Few-shot prompt (pseudo-code)
few_shot_prompt = (
    "Translate the following English sentences to French.\n"
    "Example 1: 'Good morning' -> 'Bonjour'\n"
    "Example 2: 'Thank you' -> 'Merci'\n"
    "Now, translate: 'How are you?'"
)
# response = language_model_api(prompt=few_shot_prompt)
response = "Comment ça va ?"  # Example output
print("Prompt:\n", few_shot_prompt)
print("Response:", response)
```

# 4. Chain-of-Thought Prompting

Chain-of-thought prompting encourages the model to reason step-by-step by including intermediate reasoning steps in the prompt.

## Benefits:

- Improves performance on reasoning tasks
- Makes the model's thought process transparent

```python

# Example: Chain-of-thought prompt (pseudo-code)
cot_prompt = (
    "Solve the following math problem step by step.\n"
    "Question: If you have 3 apples and get 2 more, how many apples do you have?\n"
    "Let's think step by step.\n"
    "Step 1: Start with 3 apples.\n"
    "Step 2: Get 2 more apples.\n"
    "Step 3: 3 + 2 = 5.\n"
    "Answer: 5 apples."
)
# response = language_model_api(prompt=cot_prompt)
response = "Step 1: Start with 3 apples. Step 2: Get 2 more apples. Step 3: 3 + 2 = 5. Answer: 5 apples."
print("Prompt:\n", cot_prompt)
```
